# Exploratory Data Analysis: NHIS 2022 Data

This notebook performs basic exploration of the processed CDC NHIS 2022 dataset focusing on young adults aged 20-39 with mental health co-morbidities.

```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

data_path = '../data/nhis_processed.csv'
df = pd.read_csv(data_path)
df.head()
```

### Basic Statistics

```
df.describe()
```

### Target Class Distribution (CVD Risk)

```
sns.countplot(x='label', data=df)
plt.title('Target Class Distribution')
plt.show()
```

### Mental Health Co-morbidities vs CVD Risk

```
if 'ANXEV_A' in df.columns and 'DEPEV_A' in df.columns:
    fig, ax = plt.subplots(1, 2, figsize=(12, 5))
    sns.countplot(x='ANXEV_A', hue='label', data=df, ax=ax[0])
    ax[0].set_title('Anxiety vs CVD Risk')
    sns.countplot(x='DEPEV_A', hue='label', data=df, ax=ax[1])
    ax[1].set_title('Depression vs CVD Risk')
    plt.show()
```
