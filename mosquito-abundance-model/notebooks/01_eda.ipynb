# Exploratory Data Analysis (EDA)
Mosquito Abundance modeling.

```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_parquet('../data/processed/merged_dataset.parquet')
df.head()
```

```
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
plt.hist(df['egg_count'].clip(upper=1000), bins=50, alpha=0.7)
plt.title('Distribution of Egg Counts (clipped at 1000)')
plt.xlabel('Egg Count')
plt.ylabel('Frequency')
plt.show()
```
