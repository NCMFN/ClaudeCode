# Exploratory Data Analysis: Pollution-Heat Index
This notebook provides a basic EDA over the aligned environmental health data.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Apply standard styles
plt.rcParams.update({'font.size': 11, 'axes.titlesize': 13, 'axes.labelsize': 11, 
                     'xtick.labelsize': 10, 'ytick.labelsize': 10, 
                     'figure.dpi': 300, 'savefig.dpi': 300})

df = pd.read_csv('../data/processed/phi_features.csv', parse_dates=['date'])
df.head()
```

```python
plt.figure(figsize=(12, 6))
df_ny = df[df['city'] == 'New York'].set_index('date').sort_index()
df_ny_last_year = df_ny.loc[df_ny.index >= (df_ny.index.max() - pd.DateOffset(years=1))]
sns.lineplot(data=df_ny_last_year.reset_index(), x='date', y='pm25_mean')
plt.title('PM2.5 Levels in New York (Last 1 Year)')
plt.ylabel('PM2.5 (ug/m3)')
plt.show()
```
